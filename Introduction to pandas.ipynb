```python
car_sales_missing=pd.read_csv("car-missing-data - Sheet1.csv")
car_sales_missing
```

```python
car_sales_missing["year"].mean()
```

```python
car_sales_missing["year"].fillna(car_sales_missing["year"].mean())
```

```python
car_sales_missing["year"]=car_sales_missing["year"].fillna(car_sales_missing["year"].mean())
car_sales_missing
```

```python
car_sales_missing.dropna(inplace=True)
```

```python
car_sales_missing
```

```python
car_sales_missing_dropped=car_sales_missing.dropna()
car_sales_missing_dropped
```

```python
car_sales_missing_dropped.to_csv("car-sales-missing-dropped.csv")
```

```python
# column from series
seats_column=pd.Series([5,5,5])

# new column called seats
car_sales["seats"]=seats_column
car_sales
```

```python
car_sales["seats"].fillna(5,inplace=True)
car_sales
```

```python
# column from python list
Fuel_economy=[15.0,20.5,14.5,16,18,17]
car_sales["fuel economy"]=Fuel_economy
car_sales
```

```python
odometer=[20000,40000,30000,15000,10000,25000]
car_sales["odometer (km)"]=odometer
car_sales
```

```python
car_sales["total fuel used(L)"]=car_sales["odometer (km)"]/100 * car_sales["fuel economy"]
car_sales
```

```python
# create a column from a single value
car_sales["number of wheels"]=4
car_sales
```

```python
car_sales["passed road safety"]=True
car_sales.dtypes
```
